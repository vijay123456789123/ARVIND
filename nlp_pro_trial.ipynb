```python
#removing stopwords using nltk
import nltk
from nltk.corpus import stopwords as stp
```

```python
stop_words = set(stp.words('english'))
```

```python
tokenized_data_title = [[word for word in text if word.lower() not in stop_words] for text in data_title]
```

```python
""" window size implicates that it takes n word before 
    the word and n words after the window size, min_count will skip the sentence 
    with less than 2 words, which is not required as in our data set we do not have a
    small sentence
"""
model = Word2Vec(window=6,min_count=2,workers = 4)
```

```python
model.build_vocab(data_title)
```

```python
model.epochs
```

```python
model.train(data_title, total_examples=model.corpus_count, epochs=15)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
```

```python
nor = normalize
```

```python
embedding_one = model_two.wv.vectors
```

```python
embedding_two = model_desc.wv.vectors
```

```python
embedding_gen_one = model.wv.vectors
```

```python
embedding_gen_two = model_gen_desc.wv.vectors
```

```python
embedding_nltk = [embedding_one , embedding_two]
```

```python
embedding_gen = [embedding_gen_one,embedding_gen_two]
```

```python
import numpy as np
```

```python
embedding_nltk_concat = np.concatenate(embedding_nltk, axis=0)
```

```python
embedding_gen_concat = np.concatenate(embedding_gen, axis=0)
```

```python
embedding_nltk_norm = nor(embedding_nltk_concat)
```

```python
embedding_gen_nor = nor(embedding_gen_concat)
```

```python
n_cluster = 3
```

```python
kmeans_nltk = KMeans(n_cluster, random_state=42,n_init=10)
cluster_labels_nltk = kmeans_nltk.fit_predict(embedding_nltk_norm)
```

```python
kmeans_gen = KMeans(n_cluster, random_state=42, n_init=10)
cluster_labels_gen = kmeans_gen.fit_predict(embedding_gen_nor)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
for i in range(n_cluster):
    cluster_i_indices = np.where(cluster_labels_nltk == i)[0]  # Get indices of embeddings in cluster i
    if len(cluster_i_indices) > 0:  # Check if cluster is not empty
        cluster_i_embeddings = embedding_nltk[cluster_i_indices]  # Get embeddings in cluster i
        plt.scatter(cluster_i_embeddings[:, 0], cluster_i_embeddings[:, 1], label=f"Cluster {i}")
```
